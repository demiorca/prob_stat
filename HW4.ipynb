```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

### Задание 1

```python
mu, sigma = 0, 0.1
```

```python
a = np.random.normal(mu, sigma, 10)
b = np.random.normal(mu, sigma, 100)
c = np.random.normal(mu, sigma, 1000)
d = np.random.normal(mu, sigma, 10000)
```

#### Исследование первой выборки

```python
# a
count, bins, ignored = plt.hist(a, 41, density = True)
plt.plot(bins, 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth = 2, color = 'r')
plt.show()
```

```python
# Среднее значение
np.mean(a)
```

```python
# Медиана
np.median(a)
```

```python
a = np.around(a, decimals = 2)
# Мода
stats.mode(a)[0]
```
